```python
def drawdistplot(r,c,arr=[]):
    plt.figure(figsize=(20,35))
    i=1

    for x in arr:
        plt.subplot(r,c,i)
        plt.title(x)
        sns.distplot(df[x])
        i+=1
    plt.show()
```

```python
def rmoutliers(df,arr=[]):
    for x in arr:
        upperlimit= df[x].mean() + 2*df[x].std()
        lowerlimit= df[x].mean() - 2*df[x].std()
        df=df[(df[x] <= upperlimit) & (df[x] >= lowerlimit)]
    return df
```

```python
df=rmoutliers(df,['Pregnancies','Glucose','BloodPressure','Insulin','BMI','SkinThickness','DiabetesPedigreeFunction','Age'])
```

```python
drawboxplot(4,2,['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'])
```

####**Scaling the data**

```python
from sklearn.preprocessing import StandardScaler

x= df.drop("Outcome",axis=1)
y= df["Outcome"]

gb_x= df.drop(['Pregnancies','BloodPressure','SkinThickness','Insulin'],axis=1)
gb_y= df["Outcome"]

scaler = StandardScaler()

x= scaler.fit_transform(x)
gb_x = scaler.fit_transform(gb_x)
```

####**ANN Model Building**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model= Sequential()

model.add(Dense(8, activation='relu',input_dim= 8))
model.add(Dense(16, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=["accuracy"])
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.1)
gb_xtrain ,gb_xtest,gb_ytrain,gb_ytest = train_test_split(gb_x,gb_y,test_size=0.1)
```

```python
model.fit(xtrain,ytrain,epochs=10, validation_split=0.1)
```

```python
import numpy as np

y_log=model.predict(xtest)

y_pred = np.where(y_log>0.5,1,0)
y_pred.reshape(55,)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(ytest,y_pred)
```

####**RFE(Recursive Feature Extraction)**

```python
def rfe_col(model):
    selector= RFE(estimator=model,n_features_to_select = 4)

    selector.fit(x,y)

    for i in range(len(x[0])):
        print(df.columns[i],"||", selector.support_[i],"||",selector.ranking_[i])
```

```python
def rfe(model,arr=[]):
    model_x= df.drop(arr,axis=1)
    model_y= df['Outcome']
    
    model_x= scaler.fit_transform(model_x)

    mod_xtrain,mod_xtest,mod_ytrain,mod_ytest = train_test_split(model_x,model_y,test_size=0.1)

    model.fit(mod_xtrain,mod_ytrain)
    
    mod_pred=  model.predict(mod_xtest)

    print("Accuracy Score:",accuracy_score(mod_ytest,mod_pred),"\n")
    sns.heatmap(confusion_matrix(mod_ytest,mod_pred),annot= True)
```

####**RFE for Logistic Regression**

```python
rfe_col(LogisticRegression())
```

```python
rfe(LogisticRegression(),['Pregnancies','BloodPressure','SkinThickness','Insulin','Outcome'])
```

####**RFE for Decision Tree**

```python
rfe_col(DecisionTreeClassifier())
```

```python
rfe(DecisionTreeClassifier(),['Pregnancies','BloodPressure','SkinThickness','Insulin','Outcome'])
```

####**RFE for Random Forest**

```python
rfe_col(RandomForestClassifier())
```

```python
rfe(RandomForestClassifier(),['Pregnancies','BloodPressure','SkinThickness','Insulin','Outcome'])
```
